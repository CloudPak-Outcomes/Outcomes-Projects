## Rows per table
You can find the names of the tables by looking at the schema file.

The next cell is provided as an example on how to access the Db2 database.<br/>
Note that all thedata is in the schema `INSURANCE`

```python
# Rows per table
tables = ["INSURANCE_HOUSEHOLD", "INSURANCE_DRIVER", "INSURANCE_POLICY", 
          "INSURANCE_CLAIM", "INSURANCE_CLAIM_DETAIL", "INSURANCE_CLAIM_STATUS"]

stmt = """
SELECT COUNT(*) as nb_rows FROM INSURANCE.{}
"""
for table in tables :
    data_pd = pd.read_sql(stmt.format(table), pconn)
    print("{}: {}".format(table, data_pd['NB_ROWS'][0].tolist() ))
```
